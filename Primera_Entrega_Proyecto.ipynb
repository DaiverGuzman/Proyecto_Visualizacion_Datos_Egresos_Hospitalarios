**Visualización 1:** Egresos Hospitalarios año 2021 por mes

El grafico nos muestra el total de egresos hospitalarios por mes año 2021 de una empresa del sector salud.

```python
group_Mes = Data.groupby(['No_Pacientes','Mes']).count()['Nombre_Mes']
print(group_Mes)
```

```python
(group_Mes.unstack(level=0).plot.bar())
plt.show()
```

**¿Qué?**

La base de datos utilizada para la realización de la visualización proviene de la información de los egresos hospitalarios de una empresa del sector salud año 2021, desagregada por Especialidad, convenio, entre otros, adicionalmente contiene otras variables de estudio como son el género, la edad y estancia.

Las variables seleccionadas para la gráfica son:

No. Registros: Variable ordenada cuantitativa, esta indica el número de pacientes.

Mes: Variable de tipo cualitativa (ordenada y secuencial) del periodo del año 2021.

**¿Por Qué?**

Comparar tendencia:
Se quiere identificar los meses donde más egresan los pacientes de hospitalización en esta empresa del sector salud.

**¿Como?**

Se realiza una codificación ordenada por mes analizar la tendencia de los egresos hospitalarios del año 2021.

**Marcas y canales**

Las marcas y los canales utilizados en las visualizaciones son:

*   Marca de área sobre el total de los egresos hospitalarios por mes.
*   Canal de posición espacial vertical para representar la cantidad egresos hospitalarios.
*   Canal de posición espacial horizontal para representar los meses de medición.

**Visualización 3:** Tipo de Régimen de los pacientes Egresados de Hospitalización año 2021


El grafico nos muestra el total de egresos hospitalarios por mes año 2021 agrupado por tipo de régimen de una empresa del sector salud localidad Usaquén.

```python
import matplotlib.pyplot as plt

tab2 = Data['Tipo de Regimen'].value_counts().rename_axis('label').reset_index(name='No_Pacientes')
tab2
```

```python
fig1, ax1 = plt.subplots()

ax1.pie(tab2['No_Pacientes'], 
        labels=tab2['label'],
        autopct='%1.2f%%',
        shadow=False,
        )
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

**¿Qué?**

La base de datos utilizada para la realización de la visualización proviene de la información de los egresos hospitalarios de una empresa del sector salud año 2021, agrupada por Tipo de régimen.

Variable de Tipo categórica.

**¿Por Qué?**

Comparar tendencia:
Se quiere identificar la cantidad de egresos hospitalarios por Tipo de régimen de los pacientes de hospitalización en esta empresa del sector salud, para identificar la población que más pacientes atiende.

**¿Como?**

Utilizamos la variable Convenio para clasificar los egresos hospitalarios y agrupar por Tipo de Régimen.

**Marcas y Canales**

Las marcas y los canales utilizados en la visualización son:

*   Marca de área sobre el total de los egresos hospitalarios agrupado por Tipo de régimen.
*   Canal de posición espacial angular.
*   Canal de color para separar los diferentes tipos de régimen.
